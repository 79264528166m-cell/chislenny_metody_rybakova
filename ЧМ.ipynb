```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Функция из уравнения: 1.2 - ln(x) - 4sin(2x) = 0
def f(x):
    return 1.2 - math.log(x) - 4 * math.sin(2 * x)
```

### Задание 1

```python
print("Графический метод")
x = np.linspace(0.5, 4, 200)
y1 = np.log(x)
y2 = 1.2 - 4 * np.sin(2 * x)

plt.plot(x, y1, label='ln(x)')
plt.plot(x, y2, label='1.2 - 4sin(2x)')
plt.grid(True)
plt.legend()
plt.title('Отделение корней')
plt.show()

for i in range(len(x)-1):
    if (y1[i]-y2[i])*(y1[i+1]-y2[i+1]) < 0:
        print(f"Корень на [{x[i]:.2f}; {x[i+1]:.2f}]")
```

### Задание 2


```python
print("Метод половинного деления")
a, b, eps = 1.4, 1.6, 0.001
while b - a > eps:
    c = (a + b) / 2
    if f(a) * f(c) <= 0:
        b = c
    else:
        a = c
print(f"Корень: {(a+b)/2:.6f}")
```

### Задание 3


```python
print("Метод простой итерации")

def phi(x):
    return math.exp(1.2 - 4 * math.sin(2 * x))

x0, eps = 1.5, 1e-6
while True:
    x1 = phi(x0)
    if abs(x1 - x0) < eps:
        break
    x0 = x1
print(f"Корень: {x1:.6f}")
```
